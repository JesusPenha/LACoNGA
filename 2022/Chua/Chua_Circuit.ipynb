## Chua circuit

Plotting data from the Chua circuit with the ELVIS III testbench

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
v1 = np.loadtxt('TEK00001a.CSV', skiprows=16, delimiter=',')
v2 = np.loadtxt('TEK00002b.CSV', skiprows=16, delimiter=',')
v3 = np.loadtxt('TEK00002.CSV', skiprows=16, delimiter=',')
v4 = np.loadtxt('TEK00003.CSV', skiprows=16, delimiter=',')
v5 = np.loadtxt('TEK00004.CSV', skiprows=16, delimiter=',')
v6 = np.loadtxt('TEK00005.CSV', skiprows=16, delimiter=',')
```

```python
fig = plt.figure()
fig.set_facecolor("black")

ax = fig.add_subplot(111)
ax.patch.set_facecolor('k')
plt.plot(v1[:,1])
plt.plot(v2[:,1])
```

```python
fig = plt.figure(figsize=(8,8))
fig.set_facecolor("black")

ax = fig.add_subplot(111)
ax.patch.set_facecolor('k')
ax.patch.set_alpha(1.0)

plt.plot(v2[:,1],v1[:,1],'lightgreen')
plt.xlabel('v2 [V]')
plt.ylabel('v1 [V]')
```

```python
fig = plt.figure(figsize=(16,16))
fig.set_facecolor("black")

ax = fig.add_subplot(111)
ax.patch.set_facecolor('k')
ax.patch.set_alpha(1.0)

plt.plot(v2[:,1],v1[:,1],'lightgreen')
plt.xlabel('v2 [V]')
plt.ylabel('v1 [V]')

plt.plot(v3[:,1],v4[:,1],'orange')
plt.xlabel('v2 [V]')
plt.ylabel('v1 [V]')

plt.plot(v5[:,1],v6[:,1],'lightskyblue')
plt.xlabel('v5 [V]')
plt.ylabel('v1 [V]')
```
