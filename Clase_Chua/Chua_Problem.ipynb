### Medición de la constante de Feigenbaum

Para la medición de la constante de Feigenbaum se utilizarán los datos del archivo datos **datos_Chua_1.dat**. Al momento de hacer la medición se tomó en cuenta que los datos tomados difieren si fueron medidos subiendo el valor de la resis- tencia o si fueron tomados bajando el valor de la resistencia, esto debido a que el fenómeno contiene histéresis. Para solventar esto, se tomaron datos tanto subiendo como bajando la resistencia. En la figura 1 se muestra la gráfica de R (A.U) vs el tiempo. Se indica el punto de oscilación estable con el número 0, as ́ı como los datos medidos subiendo están marcados con el número 2 y y los datos medidos bajando estan marcados con un número 1. Los datos en la gráfica de la figura 1 se encuentran embebidos en una circunferencia para mejor identificación. De izquierda a derecha tenemos, primera, segunda y tercera bifurcación. Usted debe buscar los datos en el archivo datos para cada tramo y realizar un promedio para luego calcular la constante de Feigenbaum con su error correspondiente, recuerde que los errores solo tienen una cifra significativa. Como sabe de las clases vistas el calculo sería:


$$\delta = \frac{R1_{(2)}-R1_{(1)}}{R1_{(3)}-R1_{(2)}}$$

ó

$$\delta = \frac{R2_{(2)}-R2_{(1)}}{R2_{(3)}-R2_{(2)}}$$

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = np.loadtxt('datos_Chua_1.dat')
```

```python
plt.figure(figsize=(7,5))
plt.plot(data[:,0], data[:,1], 'r')
plt.xlabel('Time [t]', fontsize=15)
plt.ylabel('Proportional to R [A.U.]', fontsize=15)
plt.text(50,0.53,'0', fontsize=15)
plt.text(220,0.55,'2$_{(1)}$', fontsize=15)
plt.text(420,0.54,'1$_{(1)}$', fontsize=15)
plt.text(510,0.575,'1$_{(2)}$', fontsize=15)
plt.text(595,0.59,'2$_{(2)}$', fontsize=15)
plt.text(685,0.585,'1$_{(3)}$', fontsize=15)
plt.text(860,0.600,'2$_{(3)}$', fontsize=15)
plt.show()
```

Gráfica del valor de la resistencia en función del tiempo para el experimento de Chua. Se encuentran demarcados los valores de la resistencia para diferentes tipos de difurcación. El número 1 indica que se está
bajando el valor de la resistencia al momento de medir, el número 2 indica que se está subiendo el valor de la resistencia al momento de medir .El sub-indice indica el número de la difurcación.
