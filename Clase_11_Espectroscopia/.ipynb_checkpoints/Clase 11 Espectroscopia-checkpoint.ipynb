### Práctica Clase 11: Espectroscopía

En esta práctica el estudiante tiene 3 archivos de datos, dos de los cuales se le dará la identidad del elemento químico. El tercer elemento deberá ser descubierto usando la calibración de los dos primeros.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Cargar los datos

```python
Cs = np.loadtxt("LA-CoNGA8CS137.dat", skiprows=10)
Na = np.loadtxt("LA-CoNGA8Na22.dat", skiprows=10)
```

```python
plt.figure(figsize = (8,7))
plt.subplot(211)
plt.plot(Cs)
plt.title("Espectro del Cesio", fontsize = 15)
plt.ylabel("Cuentas", fontsize = 15)
plt.subplot(212)
plt.plot(Na, 'g')
plt.title("Espectro del Sodio", fontsize = 15)
plt.ylabel("Cuentas", fontsize = 15)
plt.show()
```
